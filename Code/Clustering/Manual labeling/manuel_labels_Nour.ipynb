```
def find(tag:str ,international:list,france :list,economie:list,science:list,transport:list,sports:list,sante:list,environement:list,education:list,securite:list,entreprise:list,legislation:list)-> bool:

  return (tag in (international or france or economie or science or transport or sports or sante or environement or education or securite or entreprise or legislation))
```

```
def get_labels(tags:list ,international:list,france :list,economie:list,science:list,transport:list,sports:list,sante:list,environement:list,education:list,securite:list,entreprise:list,legislation:list)-> dict:

  labels={}
  for tag in tags:
    if tag in international:
      if "international" in labels.keys():
        labels['international']+=1
      else:
         labels['international']=1
    if tag in france:
      if "france" in labels.keys():
          labels['france']+=1
      else:
          labels['france']=1

    if tag in economie:
      if "economie" in labels.keys():
        labels['economie']+=1
      else:
        labels['economie']=1
      
    if tag in science:
      if "science" in labels.keys():
        labels['science']+=1
      else:
        labels['science']=1
      

    if tag in transport:
      if "transport" in labels.keys():
        labels['transport']+=1
      else:
        labels['transport']=1


    if tag in sports:
      if "sports" in labels.keys():

        labels['sports']+=1
      else:
        labels['sports']=1
        

    if tag in sante:
      if "sante" in labels.keys():
        labels['sante']+=1
      else:
        labels['sante']=1

        
    if tag in environement:
      if str(environement) in labels.keys():
        labels['environement']+=1
      else:
        labels['environement']=1
      
    if tag in education:
      if "education" in labels.keys():
        labels['education']+=1
      else:
        labels['education']=1

    if tag in securite:
      if "securite" in labels.keys():
        labels['securite']+=1
      else:
        labels['securite']=1
      
    if tag in entreprise:
        if 'entreprise' in labels.keys():
          labels['entreprise']+=1
        else:
          labels['entreprise']=1


    if tag in legislation:
        if "legislation" in labels.keys():
          labels['legislation']+=1
        else:
          labels['legislation']=1

  return labels

```

```
labels_dict={} 
for i,row in dataClean.iterrows():
  labels_dict[row["art_id"]]=get_labels(row['tags_Name'],international,france,economie,science,transport,sports,sante,environement,education,securite,entreprise,legislation)
  dataClean.loc[i,'labels']=str(labels_dict[row["art_id"]])
dataClean[['tags_Name','labels']].sample(10)
```

```
def calcul_max_similiraties(title,labels:list):
  similarityMax=0
  label_max=''
  word1 = wordnet.synsets(title)
  if len(word1) != 0:
      word1 = word1[0]
      for word2 in labels:
        word=word2
        if len(word2)!= 0:
          word2 = wordnet.synsets(word2)[0]
          similarity=word1.path_similarity(word2)
          if (similarity>similarityMax):
            similarityMax=similarity
            label_max=word
  return (label_max)
```

```
for i,row in dataClean.iterrows():
   dataClean.loc[i,'labels']=calcul_max_similiraties(row["titles"],list(labels_dict[i].keys()))

dataClean[['art_id','labels']].sample(10)
```
