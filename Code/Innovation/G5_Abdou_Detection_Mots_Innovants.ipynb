# Functions

```
liste_lexique: list = df_lexique['key_words_lemma'].tolist()

def nb_mots_detectes(article: str) -> int :

  """Documentation
  Parameters:
      article: text of the article content in the dataframe
  Out:
      cpt: number of key words content in the article
  """

  cpt: int = 0
  for elem in liste_lexique :
    if elem in article :
      cpt += 1
  return cpt
```

```
def compute_features(data: pd.DataFrame) -> pd.DataFrame:
    """Documentation
    Parameters:
        data: A dataframe with the initial content
    Out (if exists):
        data: A dataframe with new features based on the articles
    """

    data['ratio_key_words']: np.DataFrame  = data['key_words_detected']/data['nb_art_lemma']
    data['ratio_key_words']: np.DataFrame  = data['ratio_key_words'].fillna(0)

    return data 
```

```
# Count the number of time where the words in the list appear
list_key_word = df_lexique.key_words.values.tolist()

def count_key_words(text : str) -> int:
    """Documentation
    Parameters:
        text: text of the article

    Out (if exists):
        t : number of key word in  the text
    """
    t : int = 0

    if text != None:
        text = text.lower()
        text = text.split()
        for j in range(len(text)):
            if (text[j] in list_key_word):   #list_key_word: List of word that we will check in the sentences
                t += 1
            else : # For 2-word keywords
              try : # In case j+1 does not exist
                if text[j]+' '+text[j+1] in list_key_word :
                  t += 1
                else : #For 3-word keywords
                  try : # In case j+2 does not exist
                    if text[j]+' '+text[j+1]+' '+text[j+2] in list_key_word :
                      t+=1
                  except :
                    pass
              except :
                pass
    return t
```

```
def dic(term: str,dico: dict) :
  """Documentation
    Parameter:
        term: one or set of words
        dico: dictionary
  """
  if term in dico.keys() :
    dico[term] += 1
  elif term != '' :
    dico[term] = 1
```

```
list_key1: list = df_lexique['key_words_lemma'].tolist()
list_key2: list = df_lexique_gam['key_words_lemma'].tolist()
list_key: list = list_key1+list_key2

def count_words_diff(texte: str) -> int :
  """Documentation
    Parameter:
        texte: text of an article
    Out:
        len(dico): lenght of a dictionary
  """
  liste = texte.split()
  dico = {}
  fait = False
  for i in range(len(liste)) :
    try : 
      if (liste[i]+' '+liste[i+1]+' '+liste[i+2]) in list_key :
        dic(liste[i]+' '+liste[i+1]+' '+liste[i+2],dico)
        fait = True
      else :
        try :
          if (liste[i]+' '+liste[i+1]) in list_key :
            dic(liste[i]+' '+liste[i+1],dico)
            fait = True
        except :
          pass
    except :
      pass
    if not fait :
      dic(liste[i],dico)
    fait = False
  return (len(dico))
    
```
