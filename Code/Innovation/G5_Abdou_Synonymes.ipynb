# Functions

```
def synonyme(word: str) -> str:
  mot = word.replace(' ','_')
  dico_sim = {}
  try:
    w1 = wn.synsets(mot, lang='fra')[0]
    for synset in wn.synsets(mot, lang='fra'):
      for elem in synset.lemma_names('fra'):
        elem = elem.lower()
        if elem not in dico_sim.keys():
          w2 = wn.synsets(elem, lang='fra')[0]
          val = w1.wup_similarity(w2)
          if val != None:
            dico_sim[elem] = val
    k_max = ''
    v_max = 0.0
    for k,v in dico_sim.items():
      if (v > v_max and not re.match(mot+'[a-z]*',k)):
        v_max = v
        k_max = k
    return(k_max)
  except:
    return ''
  
```

```
def liste_unique(liste1: list, liste2: list) -> list :
  for elem in liste2 :
    if (elem not in liste1) and (elem != '') :
      liste1.append(elem)
  return liste1
```

```
def nettoyage_mot(liste: list) -> list :
  for i in range(len(liste)):
    liste[i] = liste[i].replace('_',' ')
  return liste
```
