## Strong/Bold words

```
text = str(data[data["art_id"] == 74]["art_content_html"].values)[2:-2]
text
```

```
# Feature 2 : Bold words

def bold_word(text: str) -> List[str]:
    """Documentation
    Parameters:
        text: Article content Html
    
    Out:
        sentences: list of bold sentences
    
    """
    soup: BeautifulSoup = BeautifulSoup(text, 'html.parser')

    sentences: List[str] = []
    for a in soup.find_all("b"):
        sentence: str = a.get_text()
        
        if len(sentence) > 1:
            sentences.append(sentence)

    return sentences

bold_word(text)
```

```
# Feature 3 : strong words

def strong_word(text: str) -> List[str]:
    """Documentation
    Parameters:
        text: Article content Html
    
    Out:
        sentences: list of strong sentences
    
    """
    soup: BeautifulSoup = BeautifulSoup(text, 'html.parser')

    sentences: List[str] = []
    for a in soup.find_all("strong"):
        sentence: str = a.get_text()
        
        if len(sentence) > 1:
            sentences.append(sentence)

    return sentences

strong_word(text)
```

## Average word count per sentence

```
text = str(data[data["art_id"] == 74]["art_content"].values)[2:-2]
text
```

```
sentences = split_sentences(text)
sentences
```

```
def average_word(sentences: List[str]) -> int:
    """Documentation
    Parameters:
        sentences: calls the function : "split_sentences"
    
    Out:
        count (int): average word count 

    """

    count_word = 0
    for sentence in sentences:
        count_word += len(sentence.split(" "))
    count_word /= len(sentences)

    return count_word

print(average_word(sentences))
```
