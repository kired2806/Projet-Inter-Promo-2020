```
list_art_content={}
list_art_html={}
for i,row in data2.iterrows():
  list_art_content[row["art_id"]]=row["art_content"]
  list_art_html[row["art_id"]]=row["art_content_html"]
```

#### split text into sentences 

```
def text_sentences(list_art_content : dict) -> dict: 
  """Documentation
    Parametres:
      list_art_content (dict): 
    Out:
      art_sentences (dict): sentences of the text and their art_id
  """
  art_sentences : dict = {}
  for id,art_content in list_art_content.items():
    sentences : list = re.split("[\.!?_]",str(art_content))
    art_sentences[id]=sentences

  return (art_sentences)
```

```
print(len(text_sentences(list_art_content)))
```

####  get important words

```
from bs4 import BeautifulSoup
```

```
def BoldWord(text: str) -> list: 
  """Documentation
    Parametres:
      text (str): 
    Out:
      sentences (list): important informations {sentences on strong/b balises}
  """
  soup: BeautifulSoup = BeautifulSoup(text, 'html.parser')
  sentences: list[str] = []
  for result in soup.find_all("strong"):
    sentence: str = result.get_text()
    if len(sentence) > 1:
      sentences.append(sentence)
  for result in soup.find_all("b"):
    sentence: str = result.get_text()
    if len(sentence) > 1:
      sentences.append(sentence)
  return sentences
    
```

```
BoldWord(str(data2[data2['art_id']==74]["art_content_html"].values)[2:-2])
```
