```
from urllib.parse import urlparse
import re
```

```
# col netloc
data["netloc"] = data["art_url"].apply(lambda x: urlparse(x).netloc)
```

```
# col netloc.com
data["netloc.com"] = data["netloc"].apply(lambda x: re.findall("\.[a-z]+", x))
```

```
# col nb_netloc.com
data["nb_netloc.com"] = data["netloc.com"].apply(lambda x: len(x))
```

```
# col path
data["path"] = data["art_url"].apply(lambda x: urlparse(x).path)
```

```
# col nb_word_path
data["nb_word_path"] = data["path"].apply(lambda x: len(x.replace("/", " ").replace("-", " ").replace("_", " ").split()))
```
